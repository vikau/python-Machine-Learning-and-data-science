```python
moviemat=data.pivot_table(index='userId',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of use ratings',ascending=False).head(10)
```

```python
starwars_user_ratings =moviemat['Star Wars (1977)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
#similar_t0_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```
