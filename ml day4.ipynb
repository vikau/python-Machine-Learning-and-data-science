```python
from sklearn.datasets import load_boston
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
```

```python
x=load_boston()
x.data.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x.data,x.target,test_size=.20,random_state=1)
```

```python
poly=PolynomialFeatures()
x_poly= poly.fit_transform(x_train)
```

```python
import pandas as pd
df= pd.DataFrame(x_poly.astype(int)) 
```

```python
df
```

```python
df.shape
```

```python
ridge_reg=Ridge()
```

```python
ridge_reg.fit(x_poly,y_train)
```

```python
#x_std_test=std.transform(X_test)
```

```python
x_poly_test=poly.transform(x_test)
```

```python
pred=ridge_reg.predict(x_poly_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,pred)
```

```python
print 'traning score:',ridge_reg.score(x_poly_test,y_test)
```
